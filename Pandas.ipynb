## Introductie



Pandas is een pakket in Python, dat toelaat om  om te gaan met tabellen. We kunnen deze tabellen vergelijken met een SQL tabel. Die tabel bestaat uit verschillende kolommen die data van verschillend type kunnen bevatten.  De rijen in zo een tabel stellen meestal de data voor over een persoon of item in de data set. Bijvoorbeeld voor iedere persoon in een dataset hebben we zijn geboortedatum, adres, telefoonnummer, belastbaar inkomen. Geboortedatum is van het type DATETIME adres is een STRING telefoonnummer kan een string of getal zijn, belastbaar inkomen is een geheel getal (int). Dus een tabel kan variabelen bevatten van verschillende types, in tegenstellling tot bijvoorbeed een matrix. Pandas is dan ook een aanvulling om numpy, deze library is bedoeld om matrices te manipuleren. Matrices bevatten elemenenten die wel van hetzelfde type zijn. Zo een tabel noemen we in Python een dataframe (DataFrame).

## Sorteren en selecteren

We kunnen de volgorde van de gegevens wijzigen door gebruik te maken van sorteren. Bijvoorbeeld we zouden de cereals kunnen sorteren volgens rating:

```python
cereals.sort_values('rating')
```

sort_values sorteert hier vajn laag naar hoog maar we kunnen deze volgorde omkeren, d.m.v. de ascending parameter. Bijvoorbeeld door ascending False (d.w.z. stijgend is niet waar) te zetten gaan we dus van hoog naar laag nu:

```python
cereals.sort_values('rating',ascending=False)
```

We kunnen zelfs met meerdere kolommen tegelijk sorteren. In dit geval wordt eerst gesorteerd volgens de eerste kolom vervolgens de tweede kolom enz...telkens kunnen we aangeven of het stijgend of dalend moet gesorteerd worden m.b.v. de ascending parameter. Bijvoorbeeld we sorteren eerst op suikers (sugars) en vervolgens op rating. Met rating van hoog naar laag en suikers van laag naar hoog.

```python
cereals.sort_values(['sugars','rating'],ascending=[True,False])
```

We kunnen ook een deel van de dataframe selecteren door kolommen of rijen er uit te pikken. Bijvoorbeeld we kunnen 2 kolommen er uit halen: suikers en rating:

```python
cereals[['sugars','rating']]
```

Verder kunnen we ook gaan kijken welke producten geen suiker bevatten. 

```python
cereals['sugars']==0
```

We verkrijgen dus een Pandas kolom of ook wel genoemd een Pandas Series met False en True waarden. Deze kunnen we gebruiken als selectiecriteria. We veranderen de namen van de kolommen d.m.v. rename methode, we zetten de parameter inplace op False. Hierdoor zullen de namen van de kolommen veranderen aan de linkerkant naar het Nederlands t.t.z. bij ontbijtgranen. De namen vin de tabel cereals blijven echter staan zoals we zien.

## Data groeperen en pivot tabellen
We kunnen op 2 manieren data groeperen. Vooreerst m.b.v. de groupby() methode, vervolgens zullen we ook pivottabbelen kunnen gebruiken.

### Groeperen

Stel nu dat we per producent de gemiddelde beoordeling willen kennen. We groeperen eerst alle producten per producent en nemen dan het gemiddelde van de beoordeling als volgt:

```python
ontbijtgranen.groupby('producentcode')['beoordeling'].mean()
```

```python
We kunnen hierbij groeperen over meerdere kolommmen en ook het gemiddelde berekenen over meerdere kolommen:
```

```python
ontbijtgranen.groupby(['producentcode','type'])['suikers','beoordeling'].mean()
```

We kunnen nog een stap verder gaan en verschillende statistieken berekenen door de agg (aggregate) methode te gebruiken. Bijvoorbeeld stel dat we standaardafwijking, gemiddelde en minimum en maximum willen berekenen. Voor de beoordeling van iedere producent.

```python
import numpy as np
ontbijtgranen.groupby(['producentcode'])['beoordeling'].agg([np.mean,np.min,np.max,np.std])
```

### Pivot tabellen

Voor spreadsheets gebruiken we voor vele toepassingen het begrip pivot tabel. Ook in python zijn er veel opties om pivot-tabellen te gebruiken. I.p.v. te werken met groupby kunnen we ook pivot tabellen gebruiken. We zien bijna geen verschil in de output van de volgende 2 lijnen code. 


```python
ontbijtgranen.groupby('producentcode')['beoordeling'].mean()
```

```python
ontbijtgranen.pivot_table(values='beoordeling',index='producentcode')
```

Standaard neemt pivot_table telkens de gemiddelde waarde, maar we kunnen dit wijzigen door de aggfunc (=de geaggregeerde functie) te wijzigen. De functies van de numpy library kunnen we gebruiken indien we eerst numpy importeren. Dus i.p.v. 

```python
ontbijtgranen.groupby('producentcode')['beoordeling'].median()
```

```python
import numpy as np
ontbijtgranen.pivot_table(values='beoordeling',index='producentcode',aggfunc=np.median)
```

We kunnen hier verder gaan en bijvoorbeeld over 2 kolommen groeperen zoals 'producentcode' en 'type' en ook 2 statistieken laten berekenen, zoals gemiddelde, standaardafwijking (np.std) en mediaan. Merk op dat de standaardafwijking niet wordt berekend voor H-type producten. Waarom niet?

```python
import numpy as np
ontbijtgranen.pivot_table(values='beoordeling',index='producentcode',columns=['type'],
                          aggfunc=[np.mean,np.std,np.median])
```

We kunnen de NaN ook vervangen door een andere waarde, bijvoorbeeld 0 door de parameter fill_value in te vullen:

```python
import numpy as np
ontbijtgranen.pivot_table(values='beoordeling',index='producentcode',columns=['type'],
                          aggfunc=[np.mean,np.std,np.median],fill_value=0)
```

Wat doet volgende code?

```python
import numpy as np
ontbijtgranen['een']=1
ontbijtgranen.pivot_table(values='een',index='producentcode',columns=['type'],
                          fill_value=0,aggfunc=[sum],margins=True)
```

Het geeft een tabel weer met het aantal producten per producent en per type. Bijvoorbeeld K(=Kellogs) heeft 23 producten, maar enkel van het type C. N heeft 5 producten van type C en 1 van type H en we zien ook in de kolom All op de hoogte N dat er staat 6 d.w.z. de som van 5 en 1. Ook zien we in de rij All onderaan dat er 74 producten van type C zijn en slachts 3 van type H.

We merken dat er 15 producten zijn met een beoordeling van minder dan 30. De meeste (meer dan 30 ) hebben een beoordeling tussen 31 en 43.

We kunnen ook scatterplots gebruiken om de data voor te stellen. Bijvoorbeeld laat ons bekijken of er tussen suikers en beoordeling een verband is. 

```python
import matplotlib.pyplot as plt # we importeren de library standaard als plt
ontbijtgranen.plot(x='suikers',y='beoordeling',kind='scatter') #we gebruiken de bar-chart en plot
plt.show()
```
